```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import random
```

```python
#10- Se tiene información diaria de la cotización de acciones en el NYSE en el archivo

#nyse_daily.csv en el siguiente formato (symbol, date, open, measure_midday,
#measure_afternoon, close, volume).
# Por cada acción cuyo nombre está indicado en el campo
#symbol, tendremos una entrada por fecha con los valores open, measure_midday,
#measure_afternoon, y close indicando respectivamente a qué valor abrio la acción, cuál fue el
#valor que tuvo al mediodía, cual fue su valor que tuvo por la tarde y cual fue su valor al cierre
#del mercado. Asimismo en volume se indica el volumen operado ese día para esa acción.

#Por otro lado se cuenta con el archivo s&p500.csv de formato (symbol, company_name) que
#indica aquellas acciones que deben ser consideradas para calcular el índice Standard & Poor’s
#500 (S&P 500).

#Se pide calcular el valor diario del índice S&P 500, teniendo en cuenta que el mismo se calcula
#como el promedio del valor promedio de las mediciones que tuvo cada acción ese día (open,
#measure_midday, measure_afternoon, close), para las 500 acciones que se encuentran en el
#archivo s&p500.csv.
#El resultado debe estar en un dataframe de la forma (date, index_name, value). Por ejemplo,
#una entrada del mismo sería (‘2019-03-24’, ‘SP500’, ‘35.46’).
#Nota : A los efectos prácticos del ejercicio consideraremos como estadísticamente significativo
#calcular el promedio con esas pocas mediciones.
```

```python
primer_df = pd.DataFrame(columns = ['symbol', 'date','open', 'measure_midday', 'measure_afternoon', 'close', 'volume'], dtype = int)
symbol = [2, 9, 28, 6, 1, 25]
date = ['2020-12-06','2019-09-06', '2020-04-20', '2019-12-06']
open = [2017, 2017, 2016, 1999, 1999, 2015, 2015]
measure_midday = [20, 200, 280, 10, 10, 250, 250]
measure_afternoon = [200, 2000, 2800, 100, 100, 2500, 25]
close = [6969, 420, 860, 1050]
volume = [3000, 4000, 5000, 600]
company_name = ['Mateocompanys','SantiCompanys','BauCompanys','GaviotaCompanys','Argerichpitolargo']
segundo_df = pd.DataFrame(columns = ['symbol','company_name'], dtype = int)
for i in range(1000):
    primer_df = primer_df.append({
        'symbol': random.choice(symbol),
        'date': random.choice(date),
        'open': random.choice(open),
        'measure_midday': random.choice(measure_midday),
        'measure_afternoon': random.choice(measure_afternoon),
        'close': random.choice(close),
        'volume': random.choice(volume)
    }, ignore_index = True)
    if i % 2 == 0:
        segundo_df = segundo_df.append({
            'symbol': random.choice(symbol),
            'company_name': random.choice(company_name)
        }, ignore_index = True)
```

```python
#Se pide calcular el valor diario del índice S&P 500, teniendo en cuenta que el mismo se calcula
#como el promedio del valor promedio de las mediciones que tuvo cada acción ese día (open,
#measure_midday, measure_afternoon, close), para las 500 acciones que se encuentran en el
#archivo s&p500.csv.
#El resultado debe estar en un dataframe de la forma (date, index_name, value). Por ejemplo,
#una entrada del mismo sería (‘2019-03-24’, ‘SP500’, ‘35.46’).
#Nota : A los efectos prácticos del ejercicio consideraremos como estadísticamente significativo
#calcular el promedio con esas pocas mediciones.
```

```python
primer_df_filtrado = primer_df[['symbol', 'date','open', 'measure_midday', 'measure_afternoon', 'close']].dropna()
primer_df_filtrado.head(5)
```

```python
primer_df_filtrado = primer_df_filtrado.merge(segundo_df, on = 'symbol')
primer_df_filtrado.head()
```

```python
primer_df_filtrado['date'] = pd.to_datetime(primer_df_filtrado['date'],errors = 'coerce')
primer_df_filtrado.info()
```

```python
filtradisimo_bro = primer_df_filtrado.groupby(['symbol','date']).agg({'open':'mean','measure_midday':'mean','measure_afternoon':'mean','close':'mean'}).reset_index()
filtradisimo_bro.head()
```

```python
filtradisimo_bro = filtradisimo_bro.set_index('symbol')
filtradisimo_bro['promedio_bro'] = filtradisimo_bro.mean(axis = 1)
filtradisimo_bro.head()
```

```python
filtradisimo_bro= filtradisimo_bro.reset_index()
```

```python
final_filtro = filtradisimo_bro[['date','symbol','promedio_bro']]
final_filtro
```
