El dataframe (sales) lista las ventas de productos con los siguientes campos: Dia, Mes, Año,
ProductID, Importe(USD). Para un mismo dia, mes y año puede venderse n veces el mismo
producto. Por otro lado tenemos una descripción de los productos en el dataframe (products):
ProductId, Title, Category, Description. Category puede ser "Men", "Women", "Kids"
Proponer un programa en Pandas que permita:

A. Indicar los títulos de los productos de la categoría "Men" para los cuales el Importe de
venta supera el promedio mensual de ventas de los productos de la misma categoría.
(por ejemplo si el promedio de Abril de "Men" es 120 dolares y un producto se vendió en
Abril a 135 dolares lo tenemos que listar). Usar Transform.

B. Indicar el top-10 de productos que se vendieron mayor cantidad de días de forma
consecutiva.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import random

datos = pd.DataFrame(columns = ['dia', 'mes','anio', 'product_id', 'importe'], dtype = int)
dia = [2, 20, 28, 1, 1, 25, 25]
mes = [2, 8, 5, 10, 10, 5, 6]
anio = [2017, 2017, 2016, 1999, 1999, 2015, 2015]
product_id = [2, 20, 28, 1, 1, 25, 25]
importe = [2, 8, 5, 10, 10, 5, 6]
titulo = ['consolador', 'secador', 'extirpador', 'afeitadora_huevos', 'palo_del_indoro', 'jabon', 'esclavo']
categoria = ['men','women','kids']
descripcion = ['hace cosas', 'no hace cosas']
products = pd.DataFrame(columns = ['product_id', 'titulo', 'categoria'])
for i in range(100):
    datos = datos.append({
        'dia': random.choice(dia),
        'mes': random.choice(mes),
        'anio': random.choice(anio),
        'product_id': random.choice(product_id),
        'importe': random.choice(importe)  
    }, ignore_index = True)
    if i % 3 == 0:
        products = products.append({
            'product_id': random.choice(product_id),
            'titulo': random.choice(titulo),
            'categoria': random.choice(categoria),
            'descripcion': random.choice(descripcion)

        }, ignore_index = True)
```

```python
datos_filtro = datos[['mes', 'product_id', 'importe']]
datos_filtro
```

```python
productos_men = products[products['categoria'] == 'men']
productos_men
```

```python
todo = datos_filtro.merge(productos_men, on = 'product_id', how = 'inner')
todo
```

```python
todo['promedio_mes'] = todo.groupby('mes')['importe'].transform('mean')
todo
```

```python
re_filtrado = todo[todo['importe']>todo['promedio_mes']]
re_filtrado
```
