```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

%matplotlib inline
```

Creacion del dataframe a operar

```python

events = pd.DataFrame(columns = ['u_id','session_id','event','value','datetime'])
u_id =  [1,2,3,4,5,6,7,8]
session_id = [2017, 2017, 2016, 1999, 1999, 2015, 2015]
evento = ['view-product', 'checkout', 'conversion', 'select', 'buy']
value = [200, 2000, 2800, 100, 100, 2500, 25]
datetime =  ['2020-12-06 011:59:01','2019-09-06 09:55:01', '2020-04-20 04:20:42', '2019-12-06 05:45:01']


for i in range(500):
    events = events.append({
        'u_id': random.choice(u_id),
        'session_id': random.choice(session_id),
        'event': random.choice(evento),
        'value': random.choice(value),
        'datetime': random.choice(datetime),
    }, ignore_index = True)
        
```

```python
events['datetime'] = pd.to_datetime(events['datetime'], errors='coerce')
events.info()
```

#### A) 

```python
total_sessions = events['u_id'].value_counts().to_frame()
total_sessions
```

```python
total_sessions = total_sessions[ (total_sessions['u_id'] >= total_sessions['u_id'].mean()) ]
total_sessions
```

#### B)

```python
events['sessions_per_user'] = events.groupby('u_id')['event'].transform('count')
events = events[ events['sessions_per_user'] > 70 ]
events
```

```python


"""
otra forma:

events = events.set_index('event')
events = events.loc[['checkout','conversion','view-product'],:]
events = events.reset_index()
"""
events = events[ (events['event'] == 'checkout') | (events['event'] == 'conversion') | (events['event'] == 'view-product') ]
events.head()
```

```python
grouped = events.groupby(['u_id','event']).agg({'event':'count'})
grouped = grouped.unstack()
grouped.columns = [ 'checkout', 'conversion', 'view-product']
grouped
```
